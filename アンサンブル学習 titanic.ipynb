```python
import pandas as pd 
```

```python
df = pd.read_csv("train.csv")
df.head()
```

```python
df.info()
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
df["Embarked"] = df["Embarked"].fillna(df["Embarked"].mode()[0])
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

cat_features = ["Sex", "Embarked"]

for col in cat_features:
    lbl = LabelEncoder()
    df[col] = lbl.fit_transform(list(df[col].values))    
```

```python
df.head().append(df.tail())
```

```python
X = df.drop(columns = ["PassengerId", "Survived", "Name", "Ticket", "Cabin"])
y = df["Survived"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
tree = DecisionTreeClassifier().fit(X_train, y_train)
tree.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rnd_forest = RandomForestClassifier(n_estimators = 1000, max_depth = 4, random_state = 1).fit(X_train, y_train)
```

```python
rnd_forest.score(X_test, y_test)
```

```python
test_df = pd.read_csv("test.csv")
test_df.head().append(test_df.tail())
```

```python
test_df.info()
```

```python
test_df["Age"] = test_df["Age"].fillna(test_df["Age"].mean())
test_df["Fare"] = test_df["Fare"].fillna(test_df["Fare"].mean())
```

```python
test_df.info()
```

```python
for col in cat_features:
    lbl = LabelEncoder()
    test_df[col] = lbl.fit_transform(list(test_df[col].values))    
```

```python
test_df.head()
```

```python
X_pred = test_df.drop(columns = ["PassengerId",  "Name", "Ticket", "Cabin"])
ID = test_df["PassengerId"]
```

```python
prediction = rnd_forest.predict(X_pred)
prediction
```

```python
submission = pd.DataFrame({
    "PassengerId":ID,
    "Survived":prediction
})
submission.head()
```

```python
submission.to_csv("ensemble_submission.csv", index = False)
```
