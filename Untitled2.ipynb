```python
import pandas as pd
import numpy as np
```

```python
train_df = pd.read_csv("train.csv")
test_df = pd.read_csv("test.csv")
```

```python
train_df.head().append(train_df.tail())
```

```python
train_df.info()
```

```python
train_df = train_df.drop(["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis = 1)
```

```python
train_df["Age"] = train_df["Age"].fillna(train_df["Age"].mean())
train_df["Embarked"] = train_df["Embarked"].fillna(train_df["Embarked"].mode()[0])
```

```python
train_df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
data = ["Sex", "Embarked"]
for d in data:
    le = LabelEncoder()
    train_df[d] = le.fit_transform(list(train_df[d].values))
```

```python
train_df.head()
```

```python
test_df.head()
```

```python
ID = test_df["PassengerId"]
test_df = test_df.drop(["PassengerId", "Name", "Ticket", "Fare", "Cabin"], axis = 1)
```

```python
test_df.info()
```

```python
test_df["Age"] = test_df["Age"].fillna(test_df["Age"].mean())
```

```python
test_df.info()
```

```python
for d in data:
    le = LabelEncoder()
    test_df[d] = le.fit_transform(list(test_df[d].values))
```

```python
test_df.head()
```

```python
X = train_df.drop(["Survived"], axis = 1)
y = train_df["Survived"]
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 1)
```

```python
import xgboost as xgb
```

```python
params = {
    "objective" : "binary:logistic",
    "eval_metric": "auc",
    "eta":0.1,
    "max_depth" : 6,
    "colsample_bytree" : 1,
    "silent" : 1
}
dtrain = xgb.DMatrix(X_train, label = y_train)
dtest = xgb.DMatrix(X_test, label = y_test)
```

```python
model = xgb.train(params=params,
                  dtrain=dtrain,
                  num_boost_round = 100,
                  early_stopping_rounds = 10,
                  evals = [(dtest, "test")]
                 )
```

```python
pred = model.predict(xgb.DMatrix(test_df), ntree_limit = model.best_ntree_limit)
pred = np.where(pred<0.5, 0, 1)
pred
```
